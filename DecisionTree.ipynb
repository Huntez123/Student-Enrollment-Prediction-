```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("C:/Users/kahur/Downloads/student_enrollment_data_expanded (2).csv")
```

```python
data
```

```python
data = data.fillna(0)
```

```python
label_encoders = {}
categorical_columns = ['Gender', 'Socioeconomic_Status'] 
```

```python
for column in categorical_columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
```

```python
features = data.drop(['Student_ID', 'Previous_Enrollments', 'GPA'], axis=1)  
target = data['Enrolled_Next_Semester']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
importances = clf.feature_importances_
feature_names = features.columns
importance_dict = dict(zip(feature_names, importances))
print("Feature Importances:", importance_dict)
```
